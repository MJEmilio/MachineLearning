## The Multi-Dimensional Scaling (MDS) Algorithm
- Compute the matrix of squared pairwise similarities $D, D_{ij} = \| x _i - x_j\|^2$

```python
data = X
ndata = np.shape(data)[0]
ndim = np.shape(data)[1]
d = np.zeros((ndata,ndata),dtype=float)
	
	# Compute the distance matrix
	# Inefficient -- not matrices
for i in range(ndata):
	for j in range(i+1,ndata):
		for k in range(ndim):
			d[i,j] += (data[i,k] - data[j,k])**2
		d[i,j] = np.sqrt(d[i,j])
		d[j,i] = d[i,j]
            
g = d.copy()           
M = -0.5*(g**2 - np.transpose(np.sum(g*g,axis=0) * np.ones((ndata,1))/ndata) - np.ones((ndata,1))* np.sum(g*g,axis=0)/ndata + np.sum(np.sum(g*g))/ndata**2)

eval,evec = np.linalg.eig(M)
eval = np.real(eval)
ind = np.argsort(eval)
eval = np.real(np.diag(eval[ind[-1::-1]]))
evec = evec[:,ind[-1::-1]]
y = np.real(np.dot(evec,np.transpose((np.sqrt(eval)))))
plt.scatter(y[:, 0], y[:, 1], c=color, marker='.')
plt.show()
```

- Compute $J = I_N −1/N$ (where $I_N$ is the $N×N$ identity function and $N$ is the number of datapoints)

```python
J = np.eye(ndata) - np.ones((ndata, ndata))/ndata
J.shape
```

- Compute $B = -\frac{1}{2} JDJ^T$

```python
B = -J.dot(d).dot(J.T)/2
B.shape
```

- Find the $L$ largest eigenvalues $\lambda_i$ of $B$, together with the corresponding eigenvectors $e_i$

```python
evals, evecs = np.linalg.eigh(B)
evals = np.real(evals)
# Sort by eigenvalue in descending order                                                  
idx   = np.argsort(evals)[::-1]
evals = evals[idx]
evecs = evecs[:,idx]
```

```python
evals.shape
```

- Put the eigenvalues into a diagonal matrix $V$ and set the eigenvectors to be columns of matrix $P$

```python
# Compute the coordinates using positive-eigenvalued components only                      
#w, = np.where(evals > 0)
V  = np.real(np.diag(np.sqrt(evals)))
P  = evecs
Y  = P.dot(V)
```

```python
Y.shape
```

```python
trans_data = Y[:,0:2]
```

# ISOMAP 

```python
from sklearn.neighbors import NearestNeighbors, kneighbors_graph
from sklearn.utils.graph import graph_shortest_path
```

```python
K=10
n_components=2
ndata = np.shape(data)[0]
ndim = np.shape(data)[1]
d = np.zeros((ndata,ndata),dtype=float)
	
# Compute the distance matrix
	# Inefficient -- not matrices
for i in range(ndata):
	for j in range(i+1,ndata):
		for k in range(ndim):
			d[i,j] += (data[i,k] - data[j,k])**2
		d[i,j] = np.sqrt(d[i,j])
		d[j,i] = d[i,j]

	# K-nearest neighbours
indices = d.argsort()
	#notneighbours = indices[:,K+1:]
neighbours = indices[:,:K+1]
	# Alternative: epsilon
	# epsilon = 0.1
	#neighbours = where(d<=epsilon)
	#notneighbours = where(d>epsilon)

h = np.ones((ndata,ndata),dtype=float)*np.inf
for i in range(ndata):
	h[i,neighbours[i,:]] = d[i,neighbours[i,:]]

print "Dijkstra's algorithm"
q = h.copy()
for i in range(ndata):
	for j in range(ndata):
		k = np.argmin(q[i,:])
		while not(np.isinf(q[i,k])):
			q[i,k] = np.inf
			for l in neighbours[k,:]:
				possible = h[i,l] + h[l,k]
				if possible < h[i,k]:
					h[i,k] = possible
			k = np.argmin(q[i,:])
print "Comnlete"
```

```python
J = np.eye(ndata) - np.ones((ndata, ndata))/ndata
B = -J.dot(g).dot(J.T)/2

evals, evecs = np.linalg.eigh(B)
evals = np.real(evals)
# Sort by eigenvalue in descending order                                                  
idx   = np.argsort(evals)[::-1]
evals = evals[idx]
evecs = evecs[:,idx]

V  = np.diag(np.sqrt(evals))
P  = evecs
Y  = P.dot(V)

ISo = Y[:,0:2]
plt.scatter(ISo[:, 0], ISo[:, 1], c=color, marker='.')
plt.show()
```

```python
from sklearn.neighbors import NearestNeighbors, kneighbors_graph
from sklearn.utils.graph import graph_shortest_path

nbrs_ = NearestNeighbors()
nbrs_.fit(X)
kng = kneighbors_graph(nbrs_, K, mode='distance')

dist_matrix_ = graph_shortest_path(kng)
dist_matrix_.shape

G = dist_matrix_ ** 2

J = np.eye(ndata) - np.ones((ndata, ndata))/ndata
B = -J.dot(G).dot(J.T)/2

evals, evecs = np.linalg.eigh(B)
#evals = np.real(evals)
# Sort by eigenvalue in descending order                                                  
idx   = np.argsort(evals)[::-1]
evals = evals[idx]
evecs = evecs[:,idx]

V  = np.diag(np.real(np.sqrt(evals)))
P  = evecs
Y  = P.dot(V)

IS2 = Y[:,0:2]
plt.scatter(IS2[:, 0], IS2[:, 1], c=color, marker='.')
plt.show()
```
