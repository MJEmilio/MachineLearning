# Gaussian Mixture Models


The output for any particular datapoint that is input to the algorithm will be the sum of the values expected by all of the $M$ Gaussians:

\begin{equation}
f(x) = \sum_{m=1}^M \pi_m \phi(x; \mu_m, \Sigma_m)
\end{equation}

where $\phi(x; \mu_m, \Sigma_m)$ is a Gaussian function with mean $\mu_m$, and covariance matriz $\Sigma_m$, and the $\pi_m$ are weights with constraint that $\sum_{m=1}^M \pi_m = 1$.

The probability that input $x_i$ belongs to class $m$ can be written as (where a hat on a variable ($\hat{\cdot}$) means that we are estimating the value of that variable):

\begin{equation}
p(x_i \in c_m) = \frac{ \hat{\pi}_m \phi(x_i; \hat{\mu_m}; \hat{\Sigma_m}) }{ \sum_{k=1}^M \hat{\pi}_m \phi(x_i; \hat{\mu_k}; \hat{\Sigma}_k) } 
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from scipy.stats import multivariate_normal

iris = datasets.load_iris()
X = iris.data[:,:2]
Y = iris.target
Y[:] = Y+1; 
```

```python
fig = plt.figure()
plt.scatter(X[:,0],X[:,1],c=Y, edgecolors='black',marker='o')
plt.show()
```

## The Gaussian Mixture Model EM Algorithm
- Initialisation
     - Guess parameters $\theta^{(0)}$. Set the initial $\mu_m$, $\sigma_m$ and $\pi_m$ values
- Repeat until convergence:
    - (E-step) compute the expectation 
    - (M-step) estimate the new parameters


```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
from scipy.stats import multivariate_normal


class GMM:
    def __init__(self,X,K,iterations):
        self.iterations = iterations
        self.K = K
        self.X = X
        self.mu = None
        self.pi = None
        self.cov = None
        self.XY = None
        
        """ Set the initial mu, covariance and pi values"""
        self.mu = np.random.randint(min(self.X[:,0]),max(self.X[:,0]),size=(self.K,len(self.X[0]))) 
        self.cov = np.zeros((self.K,len(X[0]),len(X[0]))) 

        for dim in range(len(self.cov)):
            np.fill_diagonal(self.cov[dim],5)
        self.pi = np.ones(self.K)/self.K
        log_likelihoods = [] 
        
        self.reg_cov = 1e-6*np.identity(len(self.X[0]))
        for m,c in zip(self.mu,self.cov):
            c += self.reg_cov
            multi_normal = multivariate_normal(mean=m,cov=c)
            
        
    """Define a function which runs for iterations, iterations"""
    def run(self):
        for i in range(self.iterations):               
            """E Step"""
            gamma = np.zeros((len(self.X),len(self.cov)))
            for m,co,p,r in zip(self.mu,self.cov,self.pi,range(len(gamma[0]))):
                co+=self.reg_cov
                mn = multivariate_normal(mean=m,cov=co)
                gamma[:,r] = p*mn.pdf(self.X)/np.sum([pi_c*multivariate_normal(mean=mu_c,cov=cov_c).pdf(X) for pi_c,mu_c,cov_c in zip(self.pi,self.mu,self.cov+self.reg_cov)],axis=0)
           
            """M Step"""
            self.mu = []
            self.cov = []
            self.pi = []
            log_likelihood = []
            for c in range(len(gamma[0])):
                m_c = np.sum(gamma[:,c],axis=0)
                mu_c = (1/m_c)*np.sum(self.X*gamma[:,c].reshape(len(self.X),1),axis=0)
                self.mu.append(mu_c)
                # Calculate the covariance matrix per source based on the new mean
                self.cov.append(((1/m_c)*np.dot((np.array(gamma[:,c]).reshape(len(self.X),1)*(self.X-mu_c)).T,(self.X-mu_c)))+self.reg_cov)
                 
                self.pi.append(m_c/np.sum(gamma)) 
    def predict(self,Y):
        
        x,y = np.meshgrid(np.sort(self.X[:,0]),np.sort(self.X[:,1]))
        self.XY = np.array([x.flatten(),y.flatten()]).T
           
        # PLot the point onto the fittet gaussians
        fig3 = plt.figure()
        ax2 = fig3.add_subplot(111)
        ax2.scatter(self.X[:,0],self.X[:,1])
        for m,c in zip(self.mu,self.cov):
            multi_normal = multivariate_normal(mean=m,cov=c)
            ax2.contour(np.sort(self.X[:,0]),np.sort(self.X[:,1]),multi_normal.pdf(self.XY).reshape(len(self.X),len(self.X)),alpha=0.6)

            for y in Y:
                ax2.scatter(y[0],y[1],c='blue')
        prediction = []        

    
    
GMM = GMM(X,3,200)     
GMM.run()
GMM.predict(X)
 
    
```
